```python
std_tm = 0.03
shrk_tm = 2.4
copypasta = [shrek, curve, overthinker, rape]

def do():
    saying = random.choice(sayings)
    keyboard.send('y')
    time.sleep(std_tm)
    keyboard.write(saying)
    keyboard.send('enter')
    
def s():
    copy = random.choice(copypasta)
    for i in copy:
        if keyboard.is_pressed('shift+m') == True:
            keyboard.wait()
            break
        keyboard.send('y')
        time.sleep(std_tm)
        keyboard.write(i, restore_state_after=True)
        keyboard.send('enter')
        time.sleep(shrk_tm)

    
def csgo():
    keyboard.send('y')
    do()
    
def keys():
    keyboard.add_hotkey('shift+c', lambda:csgo())
    keyboard.add_hotkey('shift+p', lambda:s())
    keyboard.add_hotkey('shift+o', lambda:keyboard.wait())
    
keys()
```
